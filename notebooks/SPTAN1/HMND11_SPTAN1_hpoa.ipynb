```python
import pandas as pd
pd.set_option('display.max_colwidth', None) # show entire column contents, important!
from IPython.display import HTML, display
import hpotk
from hpotk.ontology import Ontology
from hpotk.ontology.load.obographs import load_ontology
from pyphetools.creation import AllelicRequirement
from pyphetools.visualization import *
import importlib.metadata
__version__ = importlib.metadata.version("pyphetools")
print(f"Using pyphetools version {__version__}")
```

```python
hpo: Ontology = load_ontology('../hp.json')
ingestor = PhenopacketIngestor(indir="phenopackets")
```

```python
ppkt_d = ingestor.get_phenopacket_dictionary()
ppkt_list = list(ppkt_d.values())
```

```python
def filter_phenopacket_disease(disease_id, ppkt):
    if len(ppkt.interpretations) != 1:
        return false
    interpretation = ppkt.interpretations[0]
    if interpretation.diagnosis is None:
        return Fasle
    disease = interpretation.diagnosis.disease
    return disease.id == disease_id

def get_filtered_disease_list(ppkt_lst, disease_id):
    matches = 0
    non_matches = 0
    matching_ppkt = list()
    for ppkt in ppkt_list:
        if filter_phenopacket_disease(disease_id=disease_id, ppkt=ppkt):
            matching_ppkt.append(ppkt)
            matches += 1
        else:
            non_matches += 1
    print(f"{matches} of {len(ppkt_lst)} phenopackets had disease {disease_id}")
    return matching_ppkt

matching_ppkt = get_filtered_disease_list(disease_id="OMIM:620528", ppkt_lst=ppkt_list)
```

```python
PMID = "PMID:31332438" # Beijer et al. (2019)
builder = HpoaTableBuilder(phenopacket_list=matching_ppkt)
builder.childhood_onset(PMID,3,11).juvenile_onset(PMID,4,11).young_adult_onset(PMID,4,11)
builder.autosomal_dominant(PMID)
# PBoth of the following had juvenile onset
builder.juvenile_onset("PMID:33206935",1,1).juvenile_onset("PMID:33578420",1,1)
hpoa_table_creator = builder.build()
```
